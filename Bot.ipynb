```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Load the tokenizer and model (you can switch models here)
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModelForSequenceClassification.from_pretrained("distilbert-base-uncased")

# Example input query
query = "I have a sore throat and a headache. What could be the problem?"

# Tokenize and encode the query
inputs = tokenizer(query, return_tensors="pt")

# Get the model's prediction (logits)
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

# Print the logits or decode the predicted class
print(logits)
```
